```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(12, 1, figsize=(16, 42))
axes = axes.flatten()

months = df_country.columns[2:14]
for i, month in enumerate(months):
    axes[i].bar(df_country['YEAR'], df_country[month], label=month, color='tab:blue')
    axes[i].set_title(month, fontsize=12)
    axes[i].grid(True, linestyle='--', alpha=0.6)
    axes[i].set_ylabel('Rainfall (mm)', fontsize=10)
    axes[i].set_xlabel('Year', fontsize=10)

fig.suptitle('Monthly Rainfall Over the Years (India)', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.show()
```

# Missing Data Analysis of region dataset

```python
df_sub_division.isnull().sum()
```

* We have some missing data, now lets check the percentage of missing data in the dataset

```python
df_sub_division.isnull().mean() * 100
```

```python
missing_rows = df_sub_division[df_sub_division['SUBDIVISION'] == 'ANDAMAN & NICOBAR ISLANDS'].isnull().any(axis=1)
df_with_missing = df_sub_division[df_sub_division['SUBDIVISION'] == 'ANDAMAN & NICOBAR ISLANDS'][missing_rows]
```

```python
df_with_missing.head(5)
```

* Seeing the low percentage of missing values, one might argue in favour of just removing those rows, but since we are dealing with time-series data, we can't do this as we will lose data regarding that particular point in time.

* Secondly imputing with mean / median is also not the way to go, since time-series data might have a 'trend' or seasonality, which might get messed up during global mean / median imputation.


```python
import plotly.express as px

for column in range(2, 14):
  columnMonth = df_sub_division.columns[column]
  fig = px.line(df_sub_division[df_sub_division['SUBDIVISION'] == 'ANDAMAN & NICOBAR ISLANDS'],
                x='YEAR', y=columnMonth)

  fig.update_xaxes(rangeslider_visible = True, rangeselector = dict(
      buttons = list([
          dict(step = 'all')
      ])
  ))

  fig.show()
```

* We can see that for andaman there are missing values in all months except January and February
